```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random as random
import os
import cv2
```

```python
def R(arr,n,size,MAX,SEED,useseed):
    if(useseed==True):
        random.seed(SEED)
    return  ( int(arr[(np.clip(n-1,0,size))])  +  ((random.randint(-MAX,MAX))))
```

```python
AmtOfData=1
Size=100
```

```python
FILE=open('data.csv','w')
FILE.write("Dataset,Time,Price\n")
FILE.close()
```

```python
for y in range(AmtOfData):
    a=np.zeros(Size)

    FILE=open('data.csv','a')
    for x in range(Size):
        a[(x)] = R(a,x,Size,10,0,False)
        FILE=open('data.csv','a')
        FILE.write('{},{},{}\n'.format(y,x,a[x]))
        FILE.close()
        plt.cla()
        plt.plot(np.resize(np.arange(Size),(max(x-1,0))),np.resize(a,(max(x-1,0))))
        plt.savefig('f{}'.format(x))


    img=cv2.imread(r'f{}.png'.format(1))
    im_height=int(img.shape[0])
    im_width=int(img.shape[1])

    video=cv2.VideoWriter('./database/Data{}.avi'.format(y), cv2.VideoWriter_fourcc('P','I','M','1'), 24, (im_width,im_height))

    for x in range(Size):
        img=cv2.imread(r'f{}.png'.format(x))
        video.write(img)
    video.release()
    print("done")
    

```
