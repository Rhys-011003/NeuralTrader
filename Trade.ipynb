```python
import numpy as np
from numpy import random as random
import matplotlib.pyplot as plt
```

```python
size=20
prices=np.zeros(size)
time=np.zeros(size)

for x in range(size):
    prices[(x)]=  (random.randint(-100,100))/100
    time[(x)]=x
   
zero= prices.mean()
one= prices.max() #
mone= prices.min()

print(prices)
print("Mean:",prices.mean())
print("Max:",prices.max())
print("Min:",prices.min())



plt.plot(time,prices )
plt.plot(time,np.full(size,zero))
m=plt.plot(time,np.full(size,one))
plt.plot(time,np.full(size,mone))

plt.show()
```
