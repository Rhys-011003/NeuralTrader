# We need to __Increase__ profit
- to do this we will first iterate over the algorithm multiple times while changing certain variables
- the changing of the variables can be done first with large steps<br>then across that maxima ,smaller and smaller steps
- to find the maxima we will plot a graph Profit vs variable changed,profit on the vertical axis
- when the slope is zero we can assume that as our maxima
- Or We could use *np.max*




```python
from matplotlib import pyplot as plt
import numpy as np
```

```python


size=23
Ax = np.linspace(0, 10, size)
Ay=np.zeros(size)
for a in range(size):
    Ay[(a)]=np.sin(Ax[(a)])

plt.plot(Ax, Ay)

plt.show()
```

# The above Graph is A Graph of Sin(x) vs x

## Lets find the max of the function

```python
np.max(Ay)
```

## Now We Have The max value but we need to find the horizontal component of that 

```python
for idx,x in enumerate(Ay): 
    if(x==np.max(Ay)):
        print("Maxima is at : " ,Ax[(idx)])
        print(idx)
    
    
```
